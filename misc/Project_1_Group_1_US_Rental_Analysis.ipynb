```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

apartment_city_data_source = "data/Apartment-List-Rent-Data-City_2019-9.csv"
apartment_city_data = pd.read_csv(apartment_city_data_source)

# Display the data table for preview
apartment_city_data.head()
```

```python
apartment_city_data.describe()
```

```python
#apartment_city_data.info()
```

```python
#lets cleanup the df and drop the _ seperators to make it easier to parse out our years using a comprehension
apartment_city_data.rename(columns={'Location_Type':'LocationType','Bedroom_Size':'BedroomSize'},inplace=True)
#lets drop the Price_ prefix from our data columns
apartment_city_data.columns = [col.strip().replace('Price_', '') for col in apartment_city_data.columns]
#Let's set the index to help with grouping and to avoid nan index values when we setup the column multindex as a split of yyyy-mm
apartment_city_data.set_index(['Location','LocationType','State','BedroomSize'],inplace=True)
#apartment_city_data.info()
apartment_city_data.head()
```

```python
#Let's create our column multindex from the split of yyyy-mm into year and month levels
apartment_city_data.columns = pd.MultiIndex.from_tuples([tuple(c.split('_'))for c in apartment_city_data.columns],names=['Year','Month'])
apartment_city_data.head()
```

```python
#apartment_city_data.info()
```

```python
#Let's thorw up a quick plot to see change over time by Size / period
apartment_city_data.query('BedroomSize == "Studio"').median().plot(rot=45, grid=True, label='Studio')
apartment_city_data.query('BedroomSize == "1br"').median().plot(rot=45, grid=True, label='1br')
apartment_city_data.query('BedroomSize == "2br"').median().plot(rot=45, grid=True, label='2br')
apartment_city_data.query('BedroomSize == "3br"').median().plot(rot=45, grid=True, label='3br')
apartment_city_data.query('BedroomSize == "4br"').median().plot(rot=45, grid=True, label='4br')
plt.legend(loc='best')
plt.show()
```
